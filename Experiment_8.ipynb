```python
import networkx as nx
import matplotlib.pyplot as plt
```

Simple Graph

```python
# Create a simple graph with equal weights
G = nx.Graph()
G.add_weighted_edges_from([
    ('A', 'B', 2),
    ('A', 'C', 18),
    ('B', 'E', 4),
    ('B', 'D', 9),
    ('C', 'E', 3),
    ('D', 'E', 6)
])

# Find and print the shortest path from node 'A' to 'E' using Dijkstra's algorithm
shortest_path = nx.dijkstra_path(G, source='A', target='E')
print(f"Shortest path in Simple Graph from A to E (using Dijkstra's): {shortest_path}")

# Draw the graph
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=16, font_color='black')
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.title('Simple Graph with Shortest Path using Dijkstra\'s Algorithm')
plt.show()
```

Directed Graph

```python
# Create a directed graph with weighted edges
DG = nx.DiGraph()
DG.add_weighted_edges_from([
    ('A', 'B', 2),
    ('A', 'C', 4),
    ('B', 'D', 1),
    ('C', 'E', 2),
    ('D', 'E', 5)
])

# Find and print the shortest path from node 'B' to 'E' using Dijkstra's algorithm
shortest_path = nx.dijkstra_path(DG, source='B', target='E')
print(f"Shortest path in Directed Graph from B to E (using Dijkstra's): {shortest_path}")

# Draw the directed graph
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(DG)
nx.draw(DG, pos, with_labels=True, node_color='lightgreen', node_size=2000, font_size=16, font_color='black', arrows=True)
edge_labels = nx.get_edge_attributes(DG, 'weight')
nx.draw_networkx_edge_labels(DG, pos, edge_labels=edge_labels)
plt.title('Directed Graph with Shortest Path using Dijkstra\'s Algorithm')
plt.show()
```

Weighted Graph

```python
# Create a weighted graph
WG = nx.Graph()
WG.add_weighted_edges_from([
    ('A', 'B', 3),
    ('A', 'C', 1),
    ('B', 'D', 2),
    ('C', 'E', 4),
    ('D', 'E', 5),
    ('D', 'C', 7)
])

# Find and print the shortest path from node 'A' to 'D' using Dijkstra's algorithm
shortest_path = nx.dijkstra_path(WG, source='A', target='D')
print(f"Shortest path in Weighted Graph from A to D (using Dijkstra's): {shortest_path}")

# Draw the weighted graph
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(WG)
nx.draw(WG, pos, with_labels=True, node_color='salmon', node_size=2000, font_size=16, font_color='black')
edge_labels = nx.get_edge_attributes(WG, 'weight')
nx.draw_networkx_edge_labels(WG, pos, edge_labels=edge_labels)
plt.title('Weighted Graph with Shortest Path using Dijkstra\'s Algorithm')
plt.show()
```

Cycle Graph

```python
# Create a cycle graph with 5 nodes
C = nx.cycle_graph(5)

# Assign different weights to the edges
weights = {
    (0, 1): 2,
    (1, 2): 3,
    (2, 3): 1,
    (3, 4): 4,
    (4, 0): 5
}

# Add the weights to the edges
for edge, weight in weights.items():
    C[edge[0]][edge[1]]['weight'] = weight

# Find and print the shortest path from node 0 to node 3 using Dijkstra's algorithm
shortest_path = nx.dijkstra_path(C, source=0, target=3)
print(f"Shortest path in Cycle Graph from node 0 to node 3 (using Dijkstra's): {shortest_path}")

# Draw the cycle graph
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(C)  # positions for all nodes
nx.draw(C, pos, with_labels=True, node_color='lightcoral', node_size=2000, font_size=16, font_color='black')
edge_labels = nx.get_edge_attributes(C, 'weight')
nx.draw_networkx_edge_labels(C, pos, edge_labels=edge_labels)
plt.title('Cycle Graph with Different Weights and Shortest Path using Dijkstra\'s Algorithm')
plt.show()
```

Directed Graph with Weighted Edges

```python
# Create a directed graph
DG = nx.DiGraph()

# Add weighted edges to the directed graph
edges_with_weights = [
    ('A', 'B', 2),
    ('A', 'C', 4),
    ('B', 'C', 1),
    ('B', 'D', 7),
    ('C', 'D', 3),
    ('D', 'E', 1),
    ('C', 'E', 5)
]

# Add edges to the graph
DG.add_weighted_edges_from(edges_with_weights)

# Find and print the shortest path from node 'A' to node 'E' using Dijkstra's algorithm
shortest_path = nx.dijkstra_path(DG, source='A', target='E')
print(f"Shortest path in Directed Weighted Graph from A to E (using Dijkstra's): {shortest_path}")

# Draw the directed weighted graph
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(DG)  # positions for all nodes
nx.draw(DG, pos, with_labels=True, node_color='lightgreen', node_size=2000, font_size=16, font_color='black', arrows=True)
edge_labels = nx.get_edge_attributes(DG, 'weight')
nx.draw_networkx_edge_labels(DG, pos, edge_labels=edge_labels)
plt.title('Directed Weighted Graph with Shortest Path using Dijkstra\'s Algorithm')
plt.show()
```
